**Importing Libraries**

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
```

**Creating Simulation**

```
mean_cusat=155
sd_cusat=5
mean_gym=185
sd_gym=5
gymperc=0.3
total_pop_size=10000
```

**Creating Two Subgroups**

```
cusat_students = np.random.normal(mean_cusat, sd_cusat, int(total_pop_size*(1-gymperc)))
cusat_at_gym = np.random.normal(mean_gym, sd_gym, int(total_pop_size*gymperc))
```

```
print(cusat_students.max(), cusat_students.min())
```

```
print(cusat_at_gym.max(), cusat_at_gym.min(),cusat_at_gym.mean())
```

**Create the population from subgroup**

```
population=np.append(cusat_students, cusat_at_gym)
population
```

```
print(population.max(), population.min(), population.mean())
```

**Plotting The Curve**

```
# Set up the figure for plotting
plt.figure(figsize=(10,12))

# Plot the cusat students only
plt.subplot(3,1,1)
sns.distplot(cusat_students)
plt.title("cusat Students Only")

# Plot the Gym Goers only
plt.subplot(3,1,2)
sns.distplot(cusat_at_gym)
plt.title("Gym Goers Only")


# Plot both groups together
plt.subplot(3,1,3)
sns.distplot(population)
plt.title("Full Population of cusat Students")
plt.axvline(x = np.mean(population))

plt.show()
%matplotlib inline
```

**What Happens if We Sample from the Entire Population?**

```
numberSamps=1000

sampSize=30
```

```
# Get the sampling distribution of the mean from whole population

mean_distribution = np.empty(numberSamps)
for i in range(numberSamps):
  random_students = np.random.choice(population,sampSize)
  mean_distribution[i]=np.mean(random_students)
```

```
mean_=np.mean(mean_distribution)
mean_
```

```

# Plot the population and the biased sampling distribution
plt.figure(figsize = (10,8))

# Plotting the population again
plt.subplot(2,1,1)
sns.distplot(population)
plt.title("Full Population of cusat Students")
plt.axvline(x = np.mean(population))
plt.xlim([140,200])

# Plotting the sampling distribution
plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the Mean Weight of Gym Goers")
plt.axvline(x = np.mean(population))
plt.axvline(x = np.mean(mean_distribution),c='red')
plt.xlim([140,200])

plt.show()
```
